```python
import numpy as np
import katpoint
import matplotlib
import pylab as pl
from datetime import timedelta
%pylab inline
```

This notebook calculates the UVW coordinates for a source on the MeerKAT telescope.  

There are 3 varibles that can be changed in the cell below:

The target Right Ascension and declination "Target_RA,Target_Dec" are the target coordinates.

The observation duration is in hours, but if it is set to special python value of None the duration will be set so that the source is simulated horizon to horizon. If you choose a different time the simulation will be centred on the Transit time of the target source and the duration will be adjusted if it is longer than the duration the source is above the horizon. For circumpolar sources the  Observation duration will be 24 Hours or less if specified. 

If the target source is not up the notebook will produce a NeverUpError with infomation about the 'Simulated_Target' transiting below the horizon.

The horizon used in the calculation in degrees. The MeerKAT has a horizon of 20 Degrees, below this is discouraged.  

```python

#The target coordinates 
Target_RA,Target_Dec = "11:50:17.7", "-57:10:56.9"

#The Simulation will be centred on the Transit time of the target source
#Observation Duration is In Hours
Observation_Duration = None #In Hours None would be from Horizon to Horizon or 24 Hours if Circumpolar 
#Observation_Duration = 4.5 # An example of a Observation_Duration of 4 and a half hours

horizon = '20' # This must be a String or it is assumed to be radians if it is a float
# horizon set to 20 degrees For MeerKAT 
```

```python
ant1= katpoint.Antenna('ant1, -30:42:39.8, 21:26:38.0, 1086.6') 
ant1.ref_observer.horizon = str(horizon )
target =katpoint.Target('Simulated_Target, radec target, %s, %s'%(Target_RA,Target_Dec))
```

```python

#How long is is above the horizon we defined at the top
start_timestamp = katpoint.Timestamp() 
start_ed = start_timestamp.to_ephem_date()
target.body.compute(ant1.ref_observer)

mid = ant1.ref_observer.next_transit(target.body,start=start_ed).datetime()

if target.body.circumpolar :
    if Observation_Duration is None :
        Observation_Duration = 24.
    print ("The source is circumpolar")
    time1 = mid - timedelta(hours=Observation_Duration/2)
    time2 = mid + timedelta(hours=Observation_Duration/2)
else :
    time1_tmp=ant1.ref_observer.next_rising(target.body,start=start_ed).datetime() # Rising
    time2_tmp=ant1.ref_observer.next_setting(target.body,start=time1_tmp).datetime() # Setting after next Rise
    delta=time2_tmp-time1_tmp
    
    if Observation_Duration is not None and delta.total_seconds()/3600. < Observation_Duration :
        print("Target source not above horizon for Observation_Duration=%.2f Hours"%Observation_Duration)
        print("Target is above horizon for %.2f Hours"%(delta.total_seconds()/3600.))
        Observation_Duration = None # Go from horizon to Horizon
    
    if Observation_Duration is None :   
        time1 = time1_tmp
        time2 = time2_tmp    
    else :
        time1 = mid - timedelta(hours=Observation_Duration/2)
        time2 = mid + timedelta(hours=Observation_Duration/2)

print("Simulation Start",time1)
print("Simulation End",time2)
delta=time2-time1
upt2=delta.total_seconds()/3600.
print ("Uptime is %.2f hours"% (upt2))
```

```python
def get_uvw(target,ants,timevalues):
    #setup empty uvw
    #Then loop through the half the antennas (n,m) 
    #and calculate u,v,w for the target and antennas n and m
    u,v,w=[],[],[]
    for n in range(np.shape(ants)[0]-1):
        for m in range(n+1,np.shape(ants)[0]):
            u_new,v_new,w_new=target.uvw(ants[n], timevalues ,ants[m])
            u.extend(u_new)
            v.extend(v_new)
            w.extend(w_new)
    return np.array(u),np.array(v),np.array(w)

def get_olduvw(target,ants,timevalues):
    #setup empty uvw
    #Then loop through the antennas (n,m) 
    #and calculate u,v,w for the target and antennas n and m
    u,v,w=[],[],[]
    for n in range(np.shape(ants)[0]):
        for m in range(np.shape(ants)[0]):
            if n != m :
                u_new,v_new,w_new=target.uvw(ants[n], timevalues ,ants[m])
                u.extend(u_new)
                v.extend(v_new)
                w.extend(w_new)
    return np.array(u),np.array(v),np.array(w)
```

```python
#define a time range "t" we want to step through
t = start_timestamp.secs + np.linspace(0, upt2 * 60. * 60., 360.) # 360 is once every 6 min 
start_timestamp = katpoint.Timestamp(str(time1)) # use a specific time as the start time (in UTC ... )
print ("Simulation start time %s"%(start_timestamp))
u,v,w = get_uvw(target,ants,t)
```
